# Plot of New England Daily Mean temperatures in deg F and deg C

To begin, we initialize the *Plots* pkg.

```julia
using Plots
```

Then, we pull the data we need for the plots. We create an array of the months, and one of the of daily mean temperatures (DMT) (Wikipedia):

```julia
months = ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
DMTFC = [[29.0,31.7,38.3,48.1,57.0,67.7,73.4,72.1,64.9,54.0,44.7,34.7],[-1.7,-0.2,3.5,8.9,14.4,19.8,23.0,22.3,18.3,12.2,7.1,1.5]];
```

We now have the necessary data for our plots. To facilitate plotting, we create the following definitions made specifically to quickly plot the necessary graphs when given a set of temperatures in deg C and deg F.

```julia
function temp_plot_F(TEMP)
    F = plot(months,TEMP[1],title="Daily Mean Temperatures in NE",label="deg F",xlabel="Months",ylabel="deg F",lw=3)
    plot(F,reuse=false)
end
```

```julia
function temp_plot_C(TEMP)
    C = plot(months,TEMP[2],title="Daily Mean Temperatures in NE",label="deg C",xlabel="Months",ylabel="deg C",lw=3)
    plot(C,reuse=false)
end
```

To create the graphs, we intialize the PlotsGR namespace and call our functions using our DMTFC variable:

```julia
gr()
plot(temp_plot_F(DMTFC))
```

```julia
plot(temp_plot_C(DMTFC))
```

A plot of the DMT in deg F and deg C is presented upon calling each respective function.
